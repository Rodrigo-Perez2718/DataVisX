Prueba model_reg

```python
from datavis.Regresion import model_reg as mr
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
help(mr)
```

Regresion lineal

```python
# Generar datos aleatorios
np.random.seed(0)
x_data = np.random.rand(100)
y_data = 2 * x_data + 1 + np.random.normal(0, 0.2, size=100)

# Crear una instancia de la clase AstroReg
astroreg = mr.AstroReg(x_data, y_data)

# Realizar regresión lineal y mostrar el resultado
astroreg.Linear_regresion(title="Regresión Lineal de Datos Aleatorios", x_label="Eje X", y_label="Eje Y")
```

Regresion logística


```python
# Generacion del set de datos
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, random_state=45)

mr.AstroReg(X,y).log_regresion(0.3,42,'prueb Regresion logistica','Eje x', 'Eje y')
```

Prueba Clustering (model_cluster)

```python
import numpy as np
import matplotlib.pyplot as plt
from datavis.Clustering import model_cluster as mc

# Generar datos aleatorios para el clustering
X_data = np.random.rand(100, 1)
Y_data = np.random.rand(100, 1)
# Crear una instancia de la clase AstroCluster
astrocluster = mc.Astrodata(X_data, Y_data)

# Realizar el clustering y mostrar el resultado
astrocluster.Kme(n_clusters=3, title="K-Means Clustering con Datos Aleatorios", x_label="Eje X", y_label="Eje Y")
```
